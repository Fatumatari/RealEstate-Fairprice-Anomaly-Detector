# Simple linear regression

```python
#import libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

#model libraries
import statsmodels.api as sm
from statsmodels.tools import add_constant
```

```python
#get the dataset 
print(sns.get_dataset_names(), end=" ")
```

```python
#load the dataset
df = sns.load_dataset("mpg")
#select only numerical columns
df= df.select_dtypes("number")

#check top 4 rows
df[:4]
```

```python
#check missing values
df.isnull().sum()
```

```python
#drop the null values if any
df.dropna(inplace=True)
#check any null values
df.isna().sum()
```

```python
#duplicates
df.duplicated().sum()
```

## Simple Linear Regresssion

we model one feature against the target variable

```python
#seperate a feature from a label
X = df["horsepower"]
y = df["mpg"]
```

```python
#add constant
x_con = add_constant(X)
```

```python
#create Ols model

model1 = sm.OLS(y, x_con).fit()
```

```python
#check model summary
model1.summary()
```

## Multiple Linear Regression

```python
#seperate features from labels
X = df.drop("mpg", axis=1) #feature
y = df["mpg"] # label

#add constant
X_con1 = add_constant(X)
```

```python
#create model
model2 = sm.OLS(y,X_con1)
model2 = model2.fit()
```

```python
#check model summary
model2.summary()
```
