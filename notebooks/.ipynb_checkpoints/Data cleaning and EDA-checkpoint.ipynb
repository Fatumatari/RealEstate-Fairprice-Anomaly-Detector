## **<h1>FairPrice Check: Real Estate Price Predictor and Anomaly Detector**

# Business Understanding

## Business Problem

The real estate market is characterized by significant information asymmetry. Buyers and renters often lack the expertise and market visibility needed to determine whether a listed property price reflects its true market value. At the same time, sellers and agents may unintentionally overprice or underprice properties due to reliance on intuition, incomplete comparable listings, or outdated market information.

These pricing inefficiencies can result in prolonged time on the market, failed negotiations, financial losses, and reduced trust in the housing market. The problem is especially pronounced in urban housing markets, where property prices vary widely based on location, size, condition, and amenities, making accurate price evaluation difficult for non-expert participants.

## Proposed Solution

This project proposes a machine learning–based pricing intelligence system that learns market pricing patterns from historical real estate data. The model estimates the expected market value of a property based on its characteristics and compares this estimate to the property’s listed price.

Based on this comparison, the system flags pricing anomalies and classifies properties as:
- **Underpriced**
- **Fairly priced**
- **Overpriced**

This approach provides an objective, data-driven benchmark for evaluating property prices, reducing reliance on subjective judgment.

## Business Objective

The objective of this project is to develop a supervised machine learning model that accurately predicts a property’s expected market value and uses this prediction to identify potential mispricing in real estate listings. The tool is designed to support informed decision-making rather than replace professional valuation.

## Business Value

This solution delivers value to multiple stakeholders:

### Buyers & Renters
Helps users identify good deals, avoid overpriced properties, and negotiate more effectively, reducing financial risk.

### Sellers & Property Owners
Assists in setting competitive prices that reduce time on market while maximizing returns.

### Real Estate Agents & PropTech Platforms
Provides objective, data-backed pricing insights that enhance credibility, user trust, and pricing transparency.

### Policy Makers & Urban Planning Analysts (Future Use)
Can be extended to analyze housing affordability, spatial price disparities, and structural inefficiencies in urban housing markets.


# Data Understanding

## Data Source

The dataset used in this project was sourced from **Kenya Property Centre**, a major online real estate listing platform in Kenya. It contains property listings for both rental and sale markets across different regions, with a strong focus on urban and peri-urban areas.

Each row in the dataset represents a single property listing.

## Dataset Structure

The dataset consists of the following key variables:

### Price Information
- **price**: The listed price of the property. For rental listings, this typically represents the monthly rent, while for sale listings it represents the asking sale price.
- **price_qualifier**: Additional context for the price (e.g., “per month”), primarily relevant for rental listings.

### Property Characteristics
- **bedrooms**: Number of bedrooms in the property.
- **bathrooms**: Number of bathrooms.
- **toilets**: Number of toilets.
- **parking**: Number of available parking spaces.
- **furnished**: Binary indicator showing whether the property is furnished.
- **serviced**: Indicates whether the property is serviced.
- **shared**: Indicates whether the property is shared with other occupants.

### Property Classification
- **category**: Indicates whether the property is *For Sale* or *For Rent*.
- **type**: Broad property type (e.g., House, Apartment).
- **sub_type**: More detailed property classification (e.g., Townhouse, Detached Duplex).

### Location Information
- **state**: County or major administrative region (e.g., Nairobi, Kajiado).
- **locality**: Town or area within the county (e.g., Embakasi, Kitengela).
- **sub_locality**: More granular neighborhood information where available.

### Listing Metadata
- **id**: Unique identifier for each listing.
- **listdate**: Date the property was listed on the platform.

## Target Variable

- **price** is the primary target variable for this project.  
  The model aims to learn market pricing patterns from property characteristics and location features in order to estimate an expected market value.

## Data Quality Considerations

Initial inspection of the dataset reveals:
- Missing values in variables such as **bathrooms**, **toilets**, **sub_type**, and **sub_locality**
- Mixed data types including numerical, categorical, binary, and datetime variables
- A wide range of price values due to the inclusion of both rental and sale listings
- Potential outliers in property prices and parking counts

These issues require cleaning, transformation, and feature engineering before modeling.

## Relevance to the Business Problem

The dataset captures key determinants of property prices in the Kenyan real estate market, particularly location, property size, and property type. However, similar properties within the same locality often display significant price variation, suggesting the presence of pricing inconsistencies. This makes the dataset well-suited for developing a pricing anomaly detection model.

### Handling Missing Numerical Data

Before performing any analysis, it is crucial to identify missing values in the dataset, especially for numerical features that will be used in modeling.

In this step, we:

1. **Count missing values** for each column using `isna().sum()`.
2. **Calculate the percentage of missing values** relative to the total number of records.
3. **Combine the counts and percentages** into a single DataFrame for easier inspection.

This allows us to flag columns with significant missing data and decide whether to fill them (imputation) or drop them.  
Proper handling of missing data ensures the model receives complete and accurate input, reducing bias and improving prediction reliability.


```python
#df.select_dtypes(exclude=['object']).isnull().sum()
count=df.isna().sum()
percent=((count/df.shape[0])*100)
null=pd.DataFrame(pd.concat([count,percent],keys=['Missing values','% Missing values'],axis=1))
null
```

### Handling Missing Values in the `toilets` Column

The toilets feature has a significant number of missing values (~62%). This column is important because the number of toilets is a key indicator of property size and amenities, which directly affects pricing.

To handle the missing data, we fill the missing values with the **median** of the column:

- **Why median?**  
  - The median is robust to outliers, which is important because some properties may have an unusually high number of toilets.  
  - Using the mean could skew the data if there are extreme values (e.g., luxury properties with many toilets).  

This approach preserves the overall distribution of the toilets feature while ensuring that the dataset contains no missing values, making it suitable for modeling.


```python
# Fill missing values with median  

df['toilets'] = df['toilets'].fillna(df['toilets'].median())
```

### Handling Missing Values in the `price` Column

The price column is our **target variable**, representing the listed price of each property. There are only 3 missing values in this column (~0.02%), which is a very small fraction of the dataset.

Because price is the variable we are trying to predict, **we do not impute these missing values**, as that could bias the model. Instead, we drop the rows with missing prices to ensure the model only learns from valid data.

This preserves data integrity and ensures reliable predictions for property price anomaly detection.


```python
# Remove all rows where the price is missing

df = df.dropna(subset=['price'])
```

### Cleaning and Standardizing the `price_qualifier` Column

The `price_qualifier` column provides context for how the listed price should be interpreted (e.g., per month, per day, per acre). To make this feature consistent and ready for modeling:

2. **Standardize similar categories**  
   - All `'per calendar month'` values are replaced with `'per month'` to reduce redundant categories.

3. **Assign meaningful values based on property category**  
   - For rental properties (`category == 'For Rent'`) with missing qualifiers, we assign `'per month'`.  
   - For sale properties (`category == 'For Sale'`) with missing qualifiers, we assign `'Sale'`.

This ensures that the `price_qualifier` column is clean, consistent, and aligned with the property type, improving data quality for machine learning.


```python
#per calendar month values are replaced with per month

df.loc[(df['price_qualifier']=='per calendar month'),'price_qualifier']='per month'
```

The dataset is divided into two separate DataFrames to handle rental and sale properties independently:

- **df_rental**: Contains only properties listed for rent.

### Handling Extreme Rental Prices

The Kenya Property Centre dataset shows that the **highest rental value in Nairobi** is around 600,000 KES, while the **lowest rental value** is approximately 5,500 KES.  

To ensure that the dataset reflects realistic market prices and to avoid extreme outliers affecting the machine learning model, we cap rental prices to a **reasonable range**. This allows us to:  

- Preserve all listings while keeping the numerical values within realistic limits.  
- Maintain the original prices for anomaly detection by storing them in a separate column (`price_raw`).  
- Improve model stability and performance by reducing the impact of extreme values.  

> Note: We cap the prices rather than removing rows entirely, as extreme rentals may still be relevant for anomaly detection.


```python
sns.violinplot(x= 'price', data=df_rental)
```

```python
df_rental.shape
```

```python
# Preserve original rental prices for anomaly detection
df_rental['price_raw'] = df_rental['price']

# Cap rental prices to realistic limits for modeling
df_rental['price'] = df_rental['price'].clip(lower=5500, upper=600000)

# Check how many rentals were originally out of these realistic bounds
out_of_range_rent = df_rental[(df_rental['price_raw'] < 5500) | (df_rental['price_raw'] > 600000)]
print("Number of rental properties outside realistic range:", out_of_range_rent.shape[0])
```

## Univariate Analysis

Univariate analysis examines each variable individually to understand its **distribution, central tendency, and spread**.  
This is an essential step before modeling or detecting anomalies because it helps us:

1. **Understand the data**: Identify typical values, ranges, and variability for each feature.  
2. **Detect anomalies**: Spot extreme values or inconsistencies even after initial cleaning or capping.  
3. **Prepare for modeling**: Decide which variables may need transformations or encoding.  

### Categorical Features
- For features like `category`, `type`, `sub_type`, `state`, `locality`, and `price_qualifier`, we use **value counts and bar plots**.  
- This shows how listings are distributed across different categories, highlighting imbalances or missing values.  

### Numerical Features
- For features like `price`, `bedrooms`, `bathrooms`, `toilets`, and `parking`, we use **histograms, boxplots, and violin plots**.  
- This helps visualize the shape of distributions, detect skewness, and observe potential outliers.

```python
# Sale capped features for modeling

sale_features = [
    'price', 'bedrooms', 'bathrooms', 'toilets',
    'month', 'year'
]
```


### Raw features for anomaly detection

### Remove weak features based on correlation analysis


```python
weak_features = ['furnished', 'shared', 'serviced', 'reception']
rental_features = [f for f in rental_features if f not in weak_features]
sale_features   = [f for f in sale_features if f not in weak_features]
```

### Create derived features

```python
# Rentals

df_combined['rental_bathrooms_per_bedroom'] = (
    df_combined.loc[df_combined['category'] == 'For Rent', 'bathrooms'] /
    df_combined.loc[df_combined['category'] == 'For Rent', 'bedrooms']
)

df_combined['rental_total_rooms'] = (
    df_combined.loc[df_combined['category'] == 'For Rent', 'bedrooms'] +
    df_combined.loc[df_combined['category'] == 'For Rent', 'bathrooms'] +
    df_combined.loc[df_combined['category'] == 'For Rent'].get('reception', 0)
)
```

```python
# Add derived features to rental feature list

rental_features += ['rental_bathrooms_per_bedroom', 'rental_total_rooms']
```

```python
# Sales

df_combined['sale_bathrooms_per_bedroom'] = (
    df_combined.loc[df_combined['category'] == 'For Sale', 'bathrooms'] /
    df_combined.loc[df_combined['category'] == 'For Sale', 'bedrooms']
)

df_combined['sale_total_rooms'] = (
    df_combined.loc[df_combined['category'] == 'For Sale', 'bedrooms'] +
    df_combined.loc[df_combined['category'] == 'For Sale', 'bathrooms'] +
    df_combined.loc[df_combined['category'] == 'For Sale'].get('reception', 0)
)
```

```python
# Add derived features to sale feature list

sale_features += ['sale_bathrooms_per_bedroom', 'sale_total_rooms']
```

### Encode categorical variables

```python
categorical_features = ['type', 'category']
df_model = pd.get_dummies(
    df_combined[rental_features + sale_features + categorical_features],
    drop_first=True
)
```

# Linear Regression as a baseline Model

```python
# Separate rental and sale data

df_rental = df_combined[df_combined['category'] == 'For Rent'].copy()
df_sale   = df_combined[df_combined['category'] == 'For Sale'].copy()
```

```python
# Features created during feature engineering
rental_features = [
    'bedrooms', 'bathrooms', 'toilets', 'parking',
    'month', 'year', 'rental_bathrooms_per_bedroom', 'rental_total_rooms'
]

sale_features = [
    'bedrooms', 'bathrooms', 'toilets', 'parking',
    'month', 'year', 'sale_bathrooms_per_bedroom', 'sale_total_rooms'
]

categorical_features = ['type']  # one-hot encoding  
```

```python
# Prepare X and y

# Rentals
X_rental = df_rental[rental_features + categorical_features].copy()
y_rental = df_rental['price'].copy()

# Sales
X_sale = df_sale[sale_features + categorical_features].copy()
y_sale = df_sale['price'].copy()
```

```python
# One-hot encode categorical variables

X_rental = pd.get_dummies(X_rental, drop_first=True)
X_sale   = pd.get_dummies(X_sale, drop_first=True)
```

```python
# Replace infinities and NaNs

X_rental.replace([np.inf, -np.inf], np.nan, inplace=True)
X_sale.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Fill missing numeric values with median

imputer = SimpleImputer(strategy='median')
X_rental = pd.DataFrame(imputer.fit_transform(X_rental), columns=X_rental.columns, index=X_rental.index)
X_sale   = pd.DataFrame(imputer.fit_transform(X_sale), columns=X_sale.columns, index=X_sale.index)
```

```python
# scale numeric features

scaler_rental = StandardScaler()
X_rental = pd.DataFrame(scaler_rental.fit_transform(X_rental), columns=X_rental.columns, index=X_rental.index)

scaler_sale = StandardScaler()
X_sale = pd.DataFrame(scaler_sale.fit_transform(X_sale), columns=X_sale.columns, index=X_sale.index)
```

```python
#Train Linear Regression models

lr_rental = LinearRegression()
lr_rental.fit(X_rental, y_rental)

lr_sale = LinearRegression()
lr_sale.fit(X_sale, y_sale)
```

```python
#Evaluate model

def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)
    return mae, rmse, r2

print("Rental Linear Regression:")
print(evaluate_model(lr_rental, X_rental, y_rental))

print("\nSale Linear Regression:")
print(evaluate_model(lr_sale, X_sale, y_sale))
```

## Linear Regression Results – Rental and Sale Prices

After splitting the data into rental and sale categories, we trained baseline **Linear Regression models** using numeric and derived features, along with one-hot encoded categorical variables (`type`). The models were evaluated using MAE, RMSE, and R².

### Rental Price Model

- **Mean Absolute Error (MAE):** 62,987 KES  
  *On average, the predicted rental price is off by ~63k KES.*

- **Root Mean Squared Error (RMSE):** 88,467 KES  
  *Typical magnitude of prediction errors, slightly higher due to outliers.*

- **R² Score:** 0.46  
  *The model explains ~46% of the variance in rental prices.*

### Sale Price Model

- **Mean Absolute Error (MAE):** 16,575,246 KES  
  *Predictions for sale prices are off by ~16.6M KES on average.*

- **Root Mean Squared Error (RMSE):** 21,672,872 KES  
  *Error magnitude influenced by extreme property prices.*

- **R² Score:** 0.40  
  *The model explains only ~40% of the variance in sale prices.*

### Interpretation

- Linear Regression provides a **baseline model**, but the performance is limited:  
  - R² values indicate that much of the variance is not captured.  
  - Errors are substantial, especially for high-priced properties.  
- This is expected since property prices are influenced by **non-linear relationships** and categorical factors not fully captured by a linear model.
